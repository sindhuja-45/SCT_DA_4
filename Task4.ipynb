```python
df.describe()
```

# Channel Distribution

```python
df['Channel_Used'].value_counts()
```

# Campaign Type count

```python
df['Channel_Used'].value_counts().plot(kind='bar')
plt.title("Campaign Count by Channel")
plt.show()
```

# Average ROI by Channel


```python
channel_roi = df.groupby('Channel_Used')['ROI'].mean().sort_values(ascending=False)
print(channel_roi)
```

```python
channel_roi.plot(kind='bar')
plt.title("Average ROI by Channel")
plt.ylabel("ROI")
plt.show()
```

# Conversion_Rate by Channel

```python
conversion = df.groupby('Channel_Used')['Conversion_Rate'].mean()
conversion.plot(kind='bar')
plt.title("Average Conversion Rate by Channel")
plt.ylabel("Conversion Rate")
plt.show()
```

# Acquisition_Cost by Channel

```python
df['Acquisition_Cost'] = df['Acquisition_Cost'].replace('[$,]', '', regex=True).astype(float)
cost = df.groupby('Channel_Used')['Acquisition_Cost'].mean()
cost.plot(kind='bar')
plt.title("Average Acquisition by Channel")
plt.ylabel("Cost")
plt.show()
```

# Clicks by Channel

```python
clicks = df.groupby('Channel_Used')['Clicks'].mean()
clicks.plot(kind='bar')
plt.title("Average Clicks by Channel")
plt.ylabel("Clicks")
plt.show()
```

# Impression by Channel

```python
impressions = df.groupby('Channel_Used')['Impressions'].mean()
impressions.plot(kind='bar')
plt.title("Average Impressions by Channel")
plt.ylabel("Impressions")
plt.show()
```

# Engagement Score Distribution

```python
engagement = df.groupby('Channel_Used')['Engagement_Score'].mean()
engagement.plot(kind='bar')
plt.title("Average Engagement by Channel")
plt.ylabel("Engagement_Score")
plt.show()
```

# Correlation Heatmap

```python
numeric_cols = [
    'ROI', 'Conversion_Rate', 'Acquisition_Cost', 'Clicks', 'Impressions', 'Engagement_Score'
]
plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Correlation between ROI and Conversion Rate

```python
correlation = df['ROI'].corr(df['Conversion_Rate'])
print(f"Correlation between ROI and Conversion_Rate: {correlation}")
```

# **Marketing Funnel Analysis**

```python
impressions = df['Impressions'].sum()
clicks = df['Clicks'].sum()
funnel = {
    'Impressions': impressions,
    'Clicks': clicks,
}
plt.bar(funnel.keys(), funnel.values())
plt.title("Marketing Funnel")
plt.show()
```
